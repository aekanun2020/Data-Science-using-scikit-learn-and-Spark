```python
# Create DataFrame with None/Null to work with examples
import pandas as pd
import numpy as np
technologies   = {
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          }
row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']
df = pd.DataFrame(technologies, index=row_labels)
print(df)
```

```python
df.info()
```

# **1. Group by Single Column.**

```python
df.groupby(by='Courses').count() # one agg. function with remain columns
```

```python
df.groupby(by='Courses')['Fee'].count() # one agg. function with a specific column ('Fee')
```

```python
df.groupby(by='Courses')['Fee'].sum() # one agg. function with a specific column ('Fee')
```

```python
df.groupby(by='Courses')['Discount'].mean() # one agg. function with a specific column ('Discount')
```

```python
# many agg. function with a specific column ('Fee')
df.groupby(by='Courses')['Fee'].aggregate({'sum','max','min'}) 
```

```python
# many agg. function with many columns ('Fee' and 'Discount')
df.groupby(by='Courses').aggregate({'Fee':'mean','Discount':['max','min']})
```

# **2. Group by Multi Columns.**

```python
df.groupby(by=['Courses','Duration']).count()
```

```python
df.groupby(by=['Courses','Duration']).sum()
```
