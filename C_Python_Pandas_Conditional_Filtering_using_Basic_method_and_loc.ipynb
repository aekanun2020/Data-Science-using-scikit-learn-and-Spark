การ Filter Rows แบบระบุเงื่อนไข ซึ่งประกอบด้วย 1. เงื่อนไขเดียว (Single Condition) และ 2. หลายเงื่อนไข (Multiple Conditions)

```python
# Create DataFrame with None/Null to work with examples
import pandas as pd
import numpy as np
technologies   = {
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          }
row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']
df = pd.DataFrame(technologies, index=row_labels)
print(df)
```

# **1. Single Condition**

```python
filter_cond = df['Discount']>1100 #Define a condition
```

```python
filter_cond
```

```python
# Row filtering using basic method with a condition

df[filter_cond] 
```

```python
# Row filtering using .loc[] with a condition
# ข้อสังเกตคือ ผลลัพธ์เหมือนกับคำสั่งก่อนหน้านี้ ที่ไม่ใช้ .loc[]

df.loc[filter_cond] 
```

```python
df[df['Discount']>1100] # Row filtering using basic method with a condition
```

```python
df.loc[df['Discount']>1100] # Row filtering using .loc[] with a condition
```

```python
filter_cond = ~(df['Discount']>1100) #Define condition with inverse
```

```python
filter_cond
```

```python
df[filter_cond] # Row filtering using basic method with a condition
```

```python
df.loc[filter_cond] # Row filtering using .loc[] with a condition
```

```python
filter_cond = df['Discount'].isin([1000,1200]) #Define a condition with .isin()
```

```python
filter_cond
```

```python
df[filter_cond] # Row filtering using basic method with a condition
```

```python
df.loc[filter_cond] # Row filtering using .loc[] with a condition
```

# **2. Multiple Conditions**

```python
# Define many conditions
filter_cond = ((df['Discount'].isin([1000,1200])) & (df['Courses']=='Spark'))
```

```python
df[filter_cond] # Row filtering using basic method with many condition
```

```python
df.loc[filter_cond] # Row filtering using .loc[] with many condition
```

```python
# Row filtering using basic method with many condition (เขียนแบบบรรทัดเดียว)
df[((df['Discount'].isin([1000,1200])) & (df['Courses']=='Spark'))]
```

```python
# Row filtering using .loc[] with many condition (เขียนแบบบรรทัดเดียว)
df.loc[((df['Discount'].isin([1000,1200])) & (df['Courses']=='Spark'))]
```
